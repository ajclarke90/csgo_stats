```python
%matplotlib inline
from csstats import weapons
```

```python
import matplotlib.pyplot as plt
```

```python
CS_MAX_RANGE = 8192
MAX_WEAPONS_TO_SHOW = 5
```

```python
for weapon in weapons.weapons[:MAX_WEAPONS_TO_SHOW]:
    fig = plt.figure()
    plt.title(weapon.name)
    axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    axes.set_xlim([0, CS_MAX_RANGE])
    data_range = range(0, CS_MAX_RANGE, 100)
    data_points = [weapon.damagerange_calc_witharmor(distance, 'chest') for distance in data_range]
    axes.plot(data_range, data_points, 'b.')
```
